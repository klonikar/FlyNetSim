```python
import pandas as pd

uav0 = pd.read_csv('D2D_000.csv',delimiter=';')
uav0.columns = ['time', 'x1', 'x2', 'y1', 'y2', 'delay', 'ns_delay', 'distance']

uav1 = pd.read_csv('D2D_001.csv', delimiter=';')
uav1.columns = ['time', 'x1', 'x2', 'y1', 'y2', 'delay', 'ns_delay', 'distance']

uavs = pd.concat([uav0, uav1])


print(uavs.head())
print(uavs.columns)
uavs = uavs.sort_values(by=['time'])

uavs.to_csv('D2D.csv', index=False)
```

```python
import matplotlib.pyplot as plt
plt.plot(uavs['distance'], uavs['delay'])
plt.xlabel('Distance (m)')
plt.ylabel('Delay (s)')
plt.title('Delay in function of distance between UAVs')
plt.show()
```

```python
plt.plot(uavs['distance'], uavs['delay'])
plt.xlabel('Distance (m)')
plt.ylabel('Delay (s)')
plt.title('Delay in function of distance between UAVs (zoom)')
plt.ylim(0, 0.06)
plt.show()
```

```python
plt.plot(uavs['distance'], uavs['ns_delay'])
plt.xlabel('Distance (m)')
plt.ylabel('Delay (ms)')
plt.title('NS Delay in function of distance between UAVs')
plt.show()
```

```python
plt.plot(uavs['distance'], uavs['ns_delay'])
plt.xlabel('Distance (m)')
plt.ylabel('Delay (ms)')
plt.title('NS Delay in function of distance between UAVs (zoom)')
plt.ylim(0, 30)
plt.show()
```

```python
plt.plot(uavs['x1'], uavs['delay'], label='x1', color='red')
plt.plot(uavs['x2'], uavs['delay'], label='x2', color='blue')
plt.xlabel('X (m)')
plt.ylabel('Delay (s)')
plt.title('Delay in function of X position')
plt.show()
```

```python
plt.plot(uavs['x1'], uavs['ns_delay'], label='x1', color='red')
plt.plot(uavs['x2'], uavs['ns_delay'], label='x2', color='blue')
plt.xlabel('X (m)') 
plt.ylabel('Delay (ms)')
plt.title('NS Delay in function of X position')
plt.show()
```
